```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = 16, 8
sns.set(style="ticks")

x = np.random.randn(30000)
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.10, .85)})

sns.boxplot(x, ax=ax_box)
sns.distplot(x, ax=ax_hist)
plt.axvline(x=-2.698, color='red', linestyle=':')
plt.axvline(x=-0.67, color='red', linestyle=':')
plt.axvline(x= 0.67, color='red', linestyle=':')
plt.axvline(x= 2.698, color='red', linestyle=':')

plt.text(-2.698,0.42,r'$-2.698\sigma$',size=15, color='red', horizontalalignment='center')
plt.text(-0.6745,0.42,r'$-0.6745\sigma$',size=15, color='red', horizontalalignment='center')
plt.text( 0.6745,0.42,r'$ 0.6745\sigma$',size=15, color='red', horizontalalignment='center')
plt.text( 2.698,0.42,r'$ 2.698\sigma$',size=15, color='red', horizontalalignment='center')

plt.text(0,0.02,r'$50%$',size=15, color='red', horizontalalignment='center')
plt.text(-1.8,0.02,r'$24.65%$',size=15, color='red', horizontalalignment='center')
plt.text(1.8,0.02,r'$24.65%$',size=15, color='red', horizontalalignment='center')
plt.text(-3.5,0.02,r'$0.35%$',size=15, color='red', horizontalalignment='center')
plt.text(3.5,0.02,r'$0.35%$',size=15, color='red', horizontalalignment='center')


ax_box.set(yticks=[])
sns.despine(ax=ax_hist)
sns.despine(ax=ax_box, left=True)
```
